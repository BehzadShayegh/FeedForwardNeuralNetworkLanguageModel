```python
import json
import matplotlib.pyplot as plt
```

```python
colors = ['blue','red','green','yellow']
for i,w in enumerate([2,3,4]) :
  train = json.load(open('h35-w'+str(w)+'-lr0.02p_train.json'))
  test = json.load(open('h35-w'+str(w)+'-lr0.02p_test.json'))
  plt.plot(train, colors[i], alpha=0.4, label='train window='+str(w)) 
  plt.plot(test, colors[i], alpha=1, label='test window='+str(w))
plt.xlabel('Epoch')
plt.ylabel('Perplexity')
plt.title('Window Effect')
plt.legend()
plt.show()
```

```python
for i,lr in enumerate([0.01,0.02,0.03,0.1]) :
  train = json.load(open('h35-w4-lr'+str(lr)+'p_train.json'))
  test = json.load(open('h35-w4-lr'+str(lr)+'p_test.json'))
  plt.plot(train, colors[i], alpha=0.4, label='train learning rate='+str(lr)) 
  plt.plot(test, colors[i], alpha=1, label='test learning rate='+str(lr))
plt.xlabel('Epoch')
plt.ylabel('Perplexity')
plt.title('Learning Rate Effect')
plt.legend()
plt.show()
```

```python
for i,h in enumerate([35,50,100,150]) :
  train = json.load(open('h'+str(h)+'-w4-lr0.02p_train.json'))
  test = json.load(open('h'+str(h)+'-w4-lr0.02p_test.json'))
  plt.plot(train, colors[i], alpha=0.4, label='train hidden size='+str(h)) 
  plt.plot(test, colors[i], alpha=1, label='test hidden size='+str(h))
plt.xlabel('Epoch')
plt.ylabel('Perplexity')
plt.title('Hidden Layer Neurons Number Effect')
plt.legend()
plt.show()
```
